# Marimekko Chart

![alt text](https://datavizproject.com/wp-content/uploads/2016/06/DVP_101_200-05.png "")

A Mekko chart (also called marimekko chart) is a two-dimensional stacked chart. In addition to the varying segment heights of a regular stacked chart, a Mekko chart also has varying column widths. Column widths are scaled such that the total width matches the desired chart width.

# Examples

```python
import pandas as pd
from statsmodels.graphics.mosaicplot import mosaic
import pylab
from itertools import product
import numpy as np
%matplotlib inline
rand = np.random.random
speaks_mul_foreign_languages = list(product(['male', 'female'], ['yes', 'no']))
index = pd.MultiIndex.from_tuples(speaks_mul_foreign_languages, names=['male', 'female'])
data = pd.Series(rand(4), index=index)
mosaic(data, gap=0.01, title='Who knows multiple foregin languages? - Mosaic Chart')
pylab.show()
```
