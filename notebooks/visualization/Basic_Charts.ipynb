# Bar Chart

```python
x = ['A', 'B', 'C']
y = [10, 5, 3]
error = [0.2, 0.1, 0.1]
plt.barchart(x,y, error)
```

```python
x = ['A', 'B', 'C']
y_list = [[10,5, 3],[2,3,5]]
y_names = ['#1', '#2']
plt.barchart_stacked(x, y_list)

```

```python
x = ['A', 'B', 'C', 'D','E','F','G']
y_list = [[10,5, 3, 3, 4, 2, 1],[7,4,8, 3, 2, 5, 9],[4,6,3, 1, 2, 6, 2], [3,7,4, 6, 2, 3, 2]]

plt.barchart_grouped(x, y_list)
```
