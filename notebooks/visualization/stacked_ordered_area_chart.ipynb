# Ribbon Chart
Also called:  Ribbon Chart

![alt text](https://datavizproject.com/wp-content/uploads/2016/06/DVP_101_200-30.png)

Stacked ordered area chart is geared towards showing rank change. Use it instead of a stacked area chart when you want to show the change of order over time. The inner categories are connected by ribbons across the columns to help you visually see how the rank changes across the columns.

A similar visualization type is the sorted stream graph

```python
## Libraries
import warnings
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline, BSpline
from matplotlib.ticker import FuncFormatter, PercentFormatter
%matplotlib inline
warnings.filterwarnings('ignore')
mpl.style.use('fivethirtyeight')


def millions(x, pos=0):
    return '%1.1fM' % (x*1e-6)

def broader(x, gap=0):
    t = np.zeros(len(x)*3)
    t_i = 0
    for i, v in enumerate(x):
        t[t_i] = v - gap
        t_i = t_i + 1
        t[t_i] = v
        t_i = t_i + 1
        t[t_i] = v + gap
        t_i = t_i + 1
        
    return t


def sorted_stream_graph(df, time, value, sort_by, category, percentage=False, smooth=True, alpha=0.7):
    _df = df
    _df['cumsum'] = _df.sort_values(by=[time, sort_by], ascending=[True, True]).groupby(time)[value].cumsum()
    _df
    fig, ax = plt.subplots(figsize=(14,10))
    #colors = [[0,0,0], [230/255,159/255,0], [86/255,180/255,233/255], [0,158/255,115/255],[213/255,94/255,0], [0,114/255,178/255], [240/255, 228/255, 66/255]]
    #colors = [[0,0,0], [230/255,159/255,0], [86/255,180/255,233/255], [0,158/255,115/255],[213/255,94/255,0], [0,114/255,178/255]]
    colors = ["#920000","#490092","#009292",[230/255,159/255,0], [86/255,180/255,233/255], [0,158/255,115/255],[213/255,94/255,0], [0,114/255,178/255]]
    #colors = ["#000000","#004949","#009292","#ff6db6","#ffb6db",
    # "#490092","#006ddb","#b66dff","#6db6ff","#b6dbff",
    # "#920000","#924900","#db6d00","#24ff24","#ffff6d"]
    #colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:brown', 'tab:grey', 'tab:pink', 'tab:olive', 'deeppink', 'steelblue', 'firebrick', 'mediumseagreen']
    color_size = len(colors)

    cat = _df[category].unique()
    
    _max = _df.groupby(time)['cumsum'].max()#.values
    
    x_all = _df[time].astype(int).unique()
    x_all.sort()
    x_gap = x_all[1] - x_all[0]
    
    margin = 0.002 * _max[-1]
    
    for i, c in enumerate(cat):
        _df1 = _df[_df[category] == c]
        _df1['min'] = _df1['cumsum'] - _df1[value]
        _df1['max'] = _df1['cumsum']
        _df1.sort_values(by=time, inplace=True)
        _df1.set_index(time, inplace=True)

        x = _df1.index.astype(int)
        y_min = _df1['min']
        y_max = _df1['max']
        
        if c == 'etc':
            color = [0,0,0]
        else:    
            color = colors[i%color_size]
            
        if percentage :
            #x= x_per
            y_min = y_min/_max
            y_max = y_max/_max
            
            #y_min.fillna(0, inplace=True)
            #y_max.fillna(0, inplace=True)
            y_min.dropna(inplace=True)
            y_max.dropna(inplace=True)

        if smooth:
            #print(x.size)

            #print(x)
            #print(y_min)
            x = broader(x, x_gap * .12)
            y_min = broader(y_min.to_numpy(), 0)
            y_max = broader(y_max.to_numpy(), 0)
                
            x_min = x.min()
            x_max = x.max()
            if x.size > 1:
                k = 3
            else:
                k = 0
                x_max = x_min + x_gap
            
            x_smooth = np.linspace(x_min, x_max, x.size * 4)
            spl_min = make_interp_spline(x, y_min, k=k)
            y_min_smooth = spl_min(x_smooth)
            spl_max = make_interp_spline(x, y_max, k=k)
            y_max_smooth = spl_max(x_smooth)                        
            
            ax.fill_between(x_smooth, y_min_smooth+margin, y_max_smooth-margin, alpha=alpha, label=c, color=color)
            #ax.fill_between(x, y_min, y_max, color=colors[i%color_size], alpha=0.8, label=c)
        else:
            ax.fill_between(x, y_min+margin, y_max-margin, color=color, alpha=alpha, label=c)

        label_pos_x = x[-1]
        label_pos_y = (y_max[-1] + y_min[-1])/2
        ax.text(label_pos_x, 
                 label_pos_y, 
                 c, fontsize=12, 
                color=color, weight='bold'
               )

        #ax.set_xticklabels(labels=x)
        #print(x)

        #ax.fill_between(_df1.index, _df1['min'], _df1['max'], color=colors[i%12], alpha=0.9, label=c)
    #ax.set_xticklabels(labels=x)
    #ax.legend(loc='upper left')
    if percentage:
        ax.yaxis.set_major_formatter(PercentFormatter(xmax=1))
    else:
        ax.yaxis.set_major_formatter(FuncFormatter(millions))
    #ax.set_xlabel('Week of Year')
    #ax.set_ylabel('Weekly Active Devices')
    #ax.set_xlim(1,43)
    ax.grid(alpha=0.0)
    #ax.set_facecolor('white')
    return fig, ax
```
